```python
import ipywidgets
import matplotlib.pyplot as plt
import numpy as np
import math

e_charge = 1.6 * pow(10, -19)
Plank = 6.63 * pow(10, -34)
lightSpeed = 3 * pow(10, 8)

def get_current(photon_desity, voltage, wavelength, work_force):
  VStop = (((Plank * lightSpeed) / (wavelength * pow(10, -9))) - (work_force * e_charge)) / e_charge
  if (voltage < VStop):
      print("Current: ", (-16.02 / VStop) * photon_desity * (voltage - VStop))
  else:
      print("Current: 0")

model_plot = ipywidgets.interactive(
    get_current, 
    photon_desity = ipywidgets.FloatSlider(value=0.5, min=0.01, max=1.0, step=0.01, description='Интенсивность:', continuous_update=True, readout=True, readout_format='.2f'),
    voltage = ipywidgets.FloatSlider(value=1.0, min=0.0, max=5.0, step=0.01, description='Напряжение, V:', continuous_update=True, readout=True, readout_format='.2f'),
    wavelength = ipywidgets.IntSlider(value=500, min=200, max=800, step=1, description='Длина волны, нм:', continuous_update=True, readout=True, readout_format='d'),
    work_force = ipywidgets.RadioButtons(options=[('Натрий', 2.28), ('Калий', 2.3), ('Медь', 4.7), ('Цинк', 4.3), ('Железо', 4.5)], description = 'Материал: ', continuous_update=True)
)
model_plot


```
